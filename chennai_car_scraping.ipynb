```python
import re 
import pandas as pd 
import numpy as np 
import requests 
from bs4 import BeautifulSoup  
```

```python
# swift
url="https://www.cars24.com/buy-used-car?f=make%3A%3D%3Amaruti%3Bmodel%3Ain%3Aswift&sort=bestmatch&serveWarrantyCount=true&gaId=1891528209.1678947207&storeCityId=5732&pinId=600001"
page=requests.get(url)
soup=BeautifulSoup(page.text) 
page 
```

```python
##year,make,model
year=[]
make=[]
model=[]
for i in soup.find_all("h2",class_="_2lmIw"):
    a=i.text
    year.append(a.split(" ")[0])
    make.append(a.split(" ")[1])
    model.append(a.split(" ")[2])

##series,transmission
series=[]
transmission=[]
for i in soup.find_all("ul",class_="_1hOnS"):
    a=i.text
    series.append(a[:3])
    transmission.append(a[3:]) 
    
#kilometers,owner,fuel
kilometers=[]
owner=[]
fuel=[]
for i in soup.find_all("ul",class_="_13yb6"):
    a=i.text 
    kilometers.append(a.split("km")[0].replace(",",""))
    owner.append(a.split("km")[1].split("Owner")[0]) 
    fuel.append(a.split("km")[1].split("Owner")[1][:-5]) 
    
#emi
emi=[]
for i in soup.find_all("div",class_="_1Em-A"):
    a=i.text
    emi.append(a.split("/")[0].replace("₹","").replace(",",""))  

#price
price=[]
for i in soup.find_all("div",class_="_18ToE"):
    a=i.text
    price.append(a.split("₹")[1].replace(",",""))  

#state,city,mileage,seating
state=["Tamil Nadu"]*20
city=["Chennai"]*20
mileage=[15]*20
seating=[5]*20  
```

```python
d={"state":state,"city":city,"year":year,"make":make,"model":model,"series":series,"transmission":transmission,"fuel":fuel,"owner":owner,"emi":emi,
   "kilometers":kilometers,"mileage":mileage,"seating":seating,"price":price}
df_maruti_swift_chn=pd.DataFrame(d) 
df_maruti_swift_chn.head() 
```
